# Hozzárendelési feladat

1. feladat: Egy vállalatnak négy gépe van, és négy olyan munka, amelyeket ezeken a gépeken kell elvégezni. Minden egyes gépre pontosan egy munkát kell kijelölni, amelyet a gép teljesen elvégez. A következő táblázat mutatja, hogy az egyes gépeknek az egyes munkákra való beállítása mennyi időt igényel.
$$
\begin{array}{l|cccc}
 \text{Idő (órában)} & \text{1. munka} & \text{2. munka} & \text{3. munka} & \text{4. munka} \\ \hline
\text{1. gép} & 20 & 13 & 15 & 11 \\
\text{2. gép} & 16 & 22 & 12 & 44  \\
\text{3. gép} & 14 & 19 & 20 & 17 \\
\text{4. gép} & 9 & 17 & 10 & 13
\end{array}
$$
A vállalat *minimalizálni* szeretné a négy munka elvégzéséhez szükséges összes beállítási időt.

* Az előadáson láttuk, hogy hogy lehet megoldani a hozzárendelési feladatot a magyar módszer segítségével.
* Ugyanakkor van lehetőségünk a feladatot speciális szállítási feladatként is megoldani, melyben
    * $n=m$, $s_i=1$ és $d_j=1$ minden $i,j=1, \dots, m$
    * és az $x_{ij}$ döntési változók binárisak.
* A szállítási feladat egészértékűségi tulajdonságának köszönhetően a $0 \leq x_{ij} \leq 1$ LP-relaxált feladatnak **van egész értékű optimális megoldása!**
* Kiegyensúlyozott feladatban ráadásul $x_{ij} \leq 1$ egyenesen következik a kínálati és keresleti pontok korlátaiból.

a) Adjuk meg a hozzárendelési feladat LP-relaxáltját, majd adjuk meg annak egy optimális megoldását is!

*Megoldás:* jelölje $x_{ij}$ változók az $i$-dik géphez ($i=1,2,3,4$) a $j$-dik munka ($j=1,2,3,4$) *rendelésének mértékét*

```python
import networkx as nx

G = nx.DiGraph()

G.add_nodes_from([
    ('g1', {"supply": 1}),
    ('g2', {"supply": 1}),
    ('g3', {"supply": 1}),
    ('g4', {"supply": 1}),
    ('m1', {"supply": -1}),
    ('m2', {"supply": -1}),
    ('m3', {"supply": -1}),
    ('m4', {"supply": -1})
])
# A keresletet értelmezzük úgy, mint negatív kínálat
# Nem is feltétlenül szükséges bevezetni hozzárendelési feladatnál, hiszen minden kínálat és kereslet 1

G.add_edge('g1', 'm1', weight = 20)
G.add_edge('g1', 'm2', weight = 13)
G.add_edge('g1', 'm3', weight = 15)
G.add_edge('g1', 'm4', weight = 11)
G.add_edge('g2', 'm1', weight = 16)
G.add_edge('g2', 'm2', weight = 22)
G.add_edge('g2', 'm3', weight = 12)
G.add_edge('g2', 'm4', weight = 44)
G.add_edge('g3', 'm1', weight = 14)
G.add_edge('g3', 'm2', weight = 19)
G.add_edge('g3', 'm3', weight = 20)
G.add_edge('g3', 'm4', weight = 17)
G.add_edge('g4', 'm1', weight = 9)
G.add_edge('g4', 'm2', weight = 17)
G.add_edge('g4', 'm3', weight = 10)
G.add_edge('g4', 'm4', weight = 13)

pos={'g1':(0,4), 'g2':(0,-4), 'g3':(2,4), 'g4':(2,-4), 'm1':(1,3), 'm2':(1,1), 'm3':(1,-1), 'm4':(1,-3)}
# gépek a sarkokban, munkák középen
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels)
```

```python
# Megoldás:
from pulp import *

munkak = ['m1','m2','m3','m4']
gepek = ['g1','g2','g3','g4']

hozzarendeles = pulp.LpProblem('hozzarendelesi_feladat', LpMinimize)

x = LpVariable.dicts('x',[(i,j) for i,j in G.edges()], lowBound = 0, cat="Continuous")
# nincs szükség bináris változókra, az egészértékűség miatt elég az LP-relaxált egy csúcsponti optimális megoldását megtalálni

hozzarendeles += lpSum(G[i][j]['weight']*x[i,j] for i,j in G.edges())

for i in gepek:
    hozzarendeles += lpSum(x[i,j] for j in munkak) == 1
    
for j in munkak:
    hozzarendeles += lpSum(x[i,j] for i in gepek) == 1

megoldas = hozzarendeles.solve()

print(LpStatus[megoldas])
if LpStatus[megoldas] == 'Optimal':
    print('Minimális hozzárendelés össz. költsége: z* = ', value(hozzarendeles.objective))
    for i,j in G.edges():
        print ("x"+str((i,j))+" =", x[i,j].varValue)
```

### Variánsok:
b) A fentebbi táblázat azt mutatja, hogy az egyes gépek egy óra alatt hány terméket képesek előállítani. A vállalat maximalizálni szeretné az egy óra alatt előállítható termékek számát!
<br>c) Visszatérve az a)-beli feladatra, rendelkezésünkre áll már egy 5. gép is, melyet az egyes munkákra $8, 19, 21$ illetve $25$ órába telik beállítani. Egyben a 2. gép alkalmatlanná válik a 3. munka elvégzésére.

```python
# Megoldás: b
hozzarendeles2 = pulp.LpProblem('max_hozzarendelesi_feladat', LpMaximize)
# Vagy maximalizálásra váltunk, vagy marad minimalizálás és a célfüggvény együtthatókat cseréljük -G[i][j]['weight']-re

x = LpVariable.dicts('x',[(i,j) for i,j in G.edges()], lowBound = 0, cat="Continuous")

hozzarendeles2 += lpSum(G[i][j]['weight']*x[i,j] for i,j in G.edges())

for i in gepek:
    hozzarendeles2 += lpSum(x[i,j] for j in munkak) == 1
    
for j in munkak:
    hozzarendeles2 += lpSum(x[i,j] for i in gepek) == 1

megoldas = hozzarendeles2.solve()

print(LpStatus[megoldas])
if LpStatus[megoldas] == 'Optimal':
    print('Maximális hozzárendelés össz. termékei: z* = ', value(hozzarendeles2.objective))
    for i,j in G.edges():
        print ("x"+str((i,j))+" =", x[i,j].varValue)
```

```python
#c jobb megoldás
from pulp import *

G.add_edge('g5', 'm1', weight = 8)
G.add_edge('g5', 'm2', weight = 19)
G.add_edge('g5', 'm3', weight = 21)
G.add_edge('g5', 'm4', weight = 25)

gepek=['g1', 'g2', 'g3', 'g4', 'g5']
munkak=['m1', 'm2', 'm3', 'm4']

hozzarendeles= LpProblem("Hozzarendeles", LpMinimize)
x=LpVariable.dicts('x', [(i,j) for i in gepek for j in munkak], lowBound=0, cat="Contiuous")

hozzarendeles+= lpSum(G[i][j]["weight"]*x[i,j] for i in gepek for j in munkak)

for i in gepek:
    hozzarendeles+= lpSum(x[i,j] for j in munkak) <= 1

for j in munkak:
    hozzarendeles+= lpSum(x[i,j] for i in gepek) == 1

hozzarendeles+= x['g2', 'm3'] == 0
    
megoldas=hozzarendeles.solve()
print(LpStatus[megoldas])
print("Minimális hozzárendelés költsége", value(hozzarendeles.objective))
for i in gepek:
    for j in munkak:
        print("x("+i+','+j+") = ",x[i,j].varValue)
```

```python
# Megoldás: c
#sima hozzarendeléses jobb

munkak = ['m1','m2','m3','m4','m5']
gepek = ['g1','g2','g3','g4','g5']

for i in gepek:
    G.add_edge(i, 'm5', weight = 0)
G.add_edge('g5', 'm1', weight = 8)
G.add_edge('g5', 'm2', weight = 19)
G.add_edge('g5', 'm3', weight = 21)
G.add_edge('g5', 'm4', weight = 25)

G.remove_edge('g2','m3')
# Az éltörlés együtt jár a hozzárendelési korlátok lentebbi módosításával
# Ugyanez helyettesíthető egy x[2,'3'] == 0 extra korláttal, vagy a G[2]['3']['weight'] kellően nagyra emelésével

hozzarendeles3 = pulp.LpProblem('hozzarendelesi_feladat', LpMinimize)

x = LpVariable.dicts('x',[(i,j) for i,j in G.edges()], lowBound = 0, cat="Continuous")

hozzarendeles3 += lpSum(G[i][j]['weight']*x[i,j] for i,j in G.edges())

for i in gepek:
    if (i != 'g2'):
        hozzarendeles3 += lpSum(x[i,j] for j in munkak) == 1
    else:
        hozzarendeles3 += lpSum(x[i,j] for j in munkak if j != 'm3') == 1
    
for j in munkak:
    if (j != 'm3'):
        hozzarendeles3 += lpSum(x[i,j] for i in gepek) == 1
    else:
        hozzarendeles3 += lpSum(x[i,j] for i in gepek if i != 'g2') == 1     

megoldas = hozzarendeles3.solve()

print(LpStatus[megoldas])
if LpStatus[megoldas] == 'Optimal':
    print('Minimális hozzárendelés össz. költsége: z* = ', value(hozzarendeles3.objective))
    for i,j in G.edges():
        print ("x"+str((i,j))+" =", x[i,j].varValue)
```
